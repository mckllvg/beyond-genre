## **Import libraries**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

## **Load Dataset**

```python
print("Loading data...")
df = pd.read_csv('/content/dataset.csv')
```

## **Drop unecessary columns **

```python
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
```

## **Select core features for clustering**

```python
features = ['valence', 'energy']
X = df[features]
```

## **Scaling**

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

K-Means Clustering

```python
print("Running K-Means clustering...")
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df['mood_cluster_id'] = kmeans.fit_predict(X_scaled)
```

## **Map numerical clusters to real human-readable "Mood" labels**

```python
cluster_means = df.groupby('mood_cluster_id')[['valence', 'energy']].mean()
print("\nCluster Center Coordinates:")
print(cluster_means)

mood_labels = {}
for cluster_id, row in cluster_means.iterrows():
    if row['valence'] >= 0.5 and row['energy'] >= 0.5:
        mood_labels[cluster_id] = 'Happy/Energetic'
    elif row['valence'] < 0.5 and row['energy'] >= 0.5:
        mood_labels[cluster_id] = 'Angry/Intense'
    elif row['valence'] < 0.5 and row['energy'] < 0.5:
        mood_labels[cluster_id] = 'Sad/Calm'
    else:
        mood_labels[cluster_id] = 'Peaceful/Chill'

# Apply the labels to our dataframe
df['mood_cluster'] = df['mood_cluster_id'].map(mood_labels)
```

## **Save the newly clustered data back to a CSV file**

```python
print("\nExporting clustered dataset...")
df.to_csv('spotify_tracks_clustered.csv', index=False)
print("Success! 'spotify_tracks_clustered.csv' is ready for download.")
```
